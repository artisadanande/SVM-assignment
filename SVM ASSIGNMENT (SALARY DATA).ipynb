```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv('Salary_train.csv')
```

```python
df1=pd.read_csv('Salary_test.csv')
```

```python
df.columns
```

```python
string_col=['workclass','education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
from sklearn import preprocessing
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
for i in string_col:
    df[i]=label_encoder.fit_transform(df[i])
    df1[i]=label_encoder.fit_transform(df1[i])
```

```python
columns=list(df.columns)
```

```python
train_x=df[columns[0:13]]
```

```python
train_x
```

```python
train_y=df[columns[13]]
```

```python
train_y
```

```python
test_x=df1[columns[0:13]]
```

```python
test_x
```

```python
test_y=df1[columns[13]]
```

```python
test_y
```

```python
train_x=df.iloc[0:500,0:13]
```

```python
train_x
```

```python
train_y=df.iloc[0:500,13]
```

```python
test_x=df1.iloc[0:300,0:13]
```

```python
test_y=df1.iloc[0:300,13]
```

```python
from sklearn.svm import SVC
```

```python
#kernel=linear
```

```python
model_linear=SVC(kernel='linear')
```

```python
model_linear.fit(train_x,train_y)
```

```python
train_pred_lin=model_linear.predict(train_x)
```

```python
test_pred_lin=model_linear.predict(test_x)
```

```python
train_lin_acc=np.mean(train_pred_lin==train_y)
```

```python
test_lin_acc=np.mean(test_pred_lin==test_y)
```

```python
train_lin_acc
```

```python
test_lin_acc
```

```python
SVC(kernel='rbf')
```

```python
model_rbf=SVC(kernel='rbf')
```

```python
model_rbf.fit(train_x,train_y)
```

```python
train_pred_rbf=model_rbf.predict(train_x)
```

```python
test_pred_rbf=model_rbf.predict(test_x)
```

```python
train_rbf_acc=np.mean(train_pred_rbf==train_y)
```
